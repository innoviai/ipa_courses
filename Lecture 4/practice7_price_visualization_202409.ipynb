```python
STOCK_CODE = '0700.HK'
START_DATE = datetime.date(2023, 1, 1)
END_DATE = datetime.date(2023, 12, 31)
```

## Fetch data from Yahoo! Finance

* Create a ticker object using `ticker = yf.Ticker(...)`
* Full documentation can be found at: https://github.com/ranaroussi/yfinance/wiki/Ticker
* `ticker.fast_info` offers a bunch of useful information like `currency`
* We can use `ticker.history` to fetch historical prices and volumes

```python
ticker = yf.Ticker(STOCK_CODE)
currency = ticker.fast_info['currency']
df = ticker.history(start=START_DATE, end=END_DATE)
print(f'Currency of {STOCK_CODE} is {currency}')
df
```
